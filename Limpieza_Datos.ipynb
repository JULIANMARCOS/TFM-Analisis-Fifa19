```python
# Importo los paquetes necesarios para empezar a trabajar con el fichero

import pandas as pd
%pylab inline

# Lectura del fichero
df = pd.read_csv("Jugadores_Fifa2019.csv")
```

```python
# Voy a eliminar las variables que no necesito para el estudio

jugadores_df = df.drop(columns=['ID','Photo','Flag','Club Logo','Work Rate','Real Face','Loaned From','Body Type',
                                     'Joined','Contract Valid Until','LS','ST','RS','LW','LF','CF','RF','RW','LAM',
                                     'CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB',
                                     'LCB','CB','RCB','RB'])

# Elimino la columna Unnamed:0, ya que dispongo de un índice.
jugadores_df = jugadores_df.drop(jugadores_df.columns[0], axis=1)
jugadores_df.head()
```

```python
jugadores_df.info() # Información de las columnas del nuevo fichero
```

```python
# Como veo que hay algunas variables que no tienen todos los valores como el club y la cláusula de rescisión,
# procedo a eliminar los jugadores que no las tienen
jugadores_df=jugadores_df.dropna()
jugadores_df.info()
```

```python
# Convertir Peso (Weight) de Libras a Kilos. Cada 2.2 Kilos es una libra
# Primero habrá que pasar la columna de Series a entero y eliminar la cadena (lbs) que sigue a la cifra del peso
def libras_kilos(val_libras):
    libras = int(val_libras.split('lbs')[0])
    peso_kg = libras/2.2
    peso_kg_redondeado = round(peso_kg, 1) #Redondeo el peso a 1 decimal
    return peso_kg_redondeado

jugadores_df['Weight_Kg'] = jugadores_df['Weight'].apply(libras_kilos)

# Elimino la columna anterior y dejo la nueva calculada
jugadores_df.drop(columns=['Weight'], inplace=True)
jugadores_df['Weight_Kg']
```

```python
# Convertir Altura (Height) que nos dan en Pies y Pulgadas a Cm. Cada 30,48 cm es un pie y 2.54 cm es una Pulgada
# La columna tiene  tipo Series y hay que pasarla a entero 
type (jugadores_df['Height'])
```

```python
# Función que coge el valor de la altura en pies y pulgadas y la convierte a cm
def pulgadas_cm(valor_pulgadas):
    pies = valor_pulgadas.split("'")[0]
    pulgadas = valor_pulgadas.split("'")[1]
    altura_cm = (int(pies) * 30.48) + (int(pulgadas)*2.54)
    altura_cm_redondeada=round(altura_cm) #Quito los valores decimales que no me aportan nada
    return (altura_cm_redondeada)
    
jugadores_df['Height_cms'] = jugadores_df['Height'].apply(pulgadas_cm)

# Elimino la columna anterior y dejo la nueva calculada en centímetros sin decimales
jugadores_df.drop(columns=['Height'], inplace=True) 
jugadores_df['Height_cms']

```
